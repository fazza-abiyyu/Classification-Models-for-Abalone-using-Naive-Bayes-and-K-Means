# **Gaussian Naive Bayes**

````python
# fetch dataset
abalone = fetch_ucirepo(id=1)

# data (as pandas dataframes)
X = abalone.data.features
y = abalone.data.targets

# metadata
print(abalone.metadata)

# variable information
print(abalone.variables)

df1 = pd.DataFrame(X)
df2 = pd.DataFrame(y)

df_1_2 = [df1, df2]
df = pd.concat(df_1_2, axis=1)
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Menampilkan informasi dataset
print(df.info())

# Menampilkan beberapa baris data pertama
print(df.head())

# Memilih fitur untuk clustering (dalam hal ini, fitur numerik kecuali Rings)
X_cluster = df.drop(['Sex', 'Rings'], axis=1)

# Standarisasi fitur untuk clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# Melatih model KMeans untuk clustering dengan k=3
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)

# Menambahkan hasil cluster ke dalam dataframe
df['Cluster'] = kmeans.labels_

# Visualisasi hasil clustering menggunakan pairplot
sns.pairplot(df, hue='Cluster', vars=X_cluster.columns)
plt.suptitle('Pairplot of Abalone Dataset by Cluster')
plt.show()

# Memilih fitur untuk klasifikasi (dalam hal ini, semua fitur numerik)
X_classification = df.drop(['Sex', 'Rings', 'Cluster'], axis=1)
y_classification = df['Cluster']

# Pisahkan data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X_classification, y_classification, test_size=0.2, random_state=42)

# Inisialisasi dan latih model Gaussian Naive Bayes
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Prediksi dengan model yang sudah dilatih
y_pred = gnb.predict(X_test)

# Evaluasi model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Classification report
print(classification_report(y_test, y_pred))

# Confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Menambahkan hasil prediksi ke dalam DataFrame
X_test_df = X_test.copy()
X_test_df['Actual Cluster'] = y_test
X_test_df['Predicted Cluster'] = y_pred

# Visualisasi hasil prediksi
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_test_df['Length'], y=X_test_df['Diameter'], hue=X_test_df['Predicted Cluster'], palette='viridis', legend='full')
plt.title('Gaussian Naive Bayes Prediction of Clusters')
plt.xlabel('Length')
plt.ylabel('Diameter')
plt.show()

# Menampilkan dataframe hasil prediksi
print("\nDataframe hasil prediksi:")
print(X_test_df.head())
````

# **K-Means**

````python
# fetch dataset
abalone = fetch_ucirepo(id=1)

# data (as pandas dataframes)
X = abalone.data.features  # Access features directly under 'data'
y = abalone.data.targets

# metadata
print(abalone.metadata)

# variable information
print(abalone.variables)

df1 = pd.DataFrame(X)
df2 = pd.DataFrame(y)

df_1_2 = [df1, df2]
df = pd.concat(df_1_2, axis=1)
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_score


# Memilih fitur untuk clustering (dalam hal ini, fitur numerik kecuali Rings)
X = df.drop(['Sex', 'Rings'], axis=1)

# Standarisasi fitur untuk clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Melakukan clustering dengan KMeans untuk mencari jumlah cluster yang optimal menggunakan metode Elbow dan Silhouette
k_values = range(2, 11)  # Coba dari 2 hingga 10 cluster
silhouette_scores = []
inertia_values = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)

    # Evaluasi menggunakan Silhouette score
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

    # Evaluasi menggunakan inertia (untuk metode Elbow)
    inertia_values.append(kmeans.inertia_)

# Plotting Silhouette scores dan metode Elbow
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Silhouette score
ax[0].plot(k_values, silhouette_scores, marker='o')
ax[0].set_title('Silhouette Score vs Number of Clusters')
ax[0].set_xlabel('Number of Clusters (k)')
ax[0].set_ylabel('Silhouette Score')
ax[0].set_xticks(k_values)
ax[0].grid(True)

# Elbow method
ax[1].plot(k_values, inertia_values, marker='o')
ax[1].set_title('Elbow Method for Optimal k')
ax[1].set_xlabel('Number of Clusters (k)')
ax[1].set_ylabel('Inertia')
ax[1].grid(True)

plt.tight_layout()
plt.show()

# Pilih nilai k berdasarkan Silhouette Score tertinggi
best_k = k_values[np.argmax(silhouette_scores)]
print(f'Best number of clusters (k): {best_k}')

# Melatih model KMeans dengan jumlah cluster terbaik
best_kmeans = KMeans(n_clusters=best_k, random_state=42)
best_kmeans.fit(X_scaled)

# Menambahkan hasil cluster ke data frame
df['Cluster'] = best_kmeans.labels_

# Menampilkan hasil clustering
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=df['Cluster'], palette='viridis', legend='full')
plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Menampilkan statistik ringkas untuk setiap cluster
cluster_summary = df.groupby('Cluster').describe() # Use describe to get summary statistics for all columns
print("\nCluster Summary:")
print(cluster_summary)
````
